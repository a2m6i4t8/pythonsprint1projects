```python
# 16. Find Top 10 Movies by Revenue

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
top_10=dataset.nlargest(10,'Revenue (Millions)')[['Title','Revenue (Millions)','Director']].set_index('Title')

# Plotting using seaborn

plt.figure(figsize=(10, 6))
sns.barplot(x=top_10.index, y='Revenue (Millions)', data=top_10)
plt.title('Top 10 Movies by Revenue (Millions)')
plt.xlabel('Movie Title')
plt.ylabel('Revenue (Millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 22. Average Rating of Movies Year-wise
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame containing movie data

data1 = dataset.groupby('Year')[['Rating']].mean().sort_values(by='Rating', ascending=False)

# Plotting using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x=data1.index, y='Rating', data=data1, palette='viridis')
plt.title('Average Rating of Movies Year-wise')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
